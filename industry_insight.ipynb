```python
import pandas as pd
```

```python
#read in the table that contains data about business confidence level in difference manufacturing sectors 
df_wide = pd.read_excel('data/Manufacturing.xlsx')
df_wide = df_wide.drop(columns='Feb-22')
df_wide = df_wide.drop([0])
df_wide
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#plot the data recorded at January 2023 in bar charts, where the red columns represent sectors that businesses have relatively low confidence in
cols = ['red' if x < 50 else 'steelblue' for x in df_wide['Jan-23']]

fig, ax = plt.subplots(figsize=(16,5))
sns.barplot(data=df_wide, x='INDUSTRIAL SECTORS', y='Jan-23', palette=cols)
plt.xticks(rotation=90)
plt.xlabel('Industrial sectors')
plt.ylabel('Business confidence index')
plt.title('Business confidence index for manufacturing industrial sectors in Brazil at Jan-23')
```

```python
#plot the data recorded at February 2023 in bar charts, where the red columns represent sectors that businesses have relatively low confidence in
cols = ['red' if x < 50 else 'steelblue' for x in df_wide['Feb-23']]

fig, ax = plt.subplots(figsize=(16,5))
sns.barplot(data=df_wide, x='INDUSTRIAL SECTORS', y='Feb-23', palette=cols)
plt.xticks(rotation=90)
plt.xlabel('Industrial sectors')
plt.ylabel('Business confidence index')
plt.title('Business confidence index for manufacturing industrial sectors in Brazil at Feb-23')
```

```python
#transform data from wide form to long form for ease of plotting
df_long = pd.melt(df_wide, id_vars='INDUSTRIAL SECTORS', value_vars=['Jan-23', 'Feb-23'])
df_long
```

```python
#plot the data recorded at January and February 2023 in side by side bar charts to see the difference in confidence index for each sectors
fig, ax = plt.subplots(figsize=(20,8))
sns.barplot(data=df_long, x='INDUSTRIAL SECTORS', y='value', hue='variable')
plt.xticks(rotation=90)
plt.xlabel('Industrial sectors')
plt.ylabel('Business confidence index')
plt.title('Business confidence index for manufacturing industrial sectors in Brazil')
```
