```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#create sample data of the current year's investment per day
series = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
investment = np.random.randint(20, 80, size=(len(series)))
df_invest = pd.DataFrame(index=series)
df_invest['investment'] = investment.tolist()
```

```python
#plot the data for investment per day in the current year
df_invest.plot(figsize=(16,5))
plt.xlabel('Months')
plt.ylabel('Investment per day in millions')
plt.title('Investment per day in the current year')
```

```python
#plot the data for investment per month in the current year
df_monthly = df_invest.groupby(by=[df_invest.index.month]).mean()

df_monthly.plot(figsize=(16,5))
plt.xlabel('Months')
plt.ylabel('Investment per month in millions')
plt.title('Investment per month in the current year')
```

```python
#create sample data of investors' share 
df_share = pd.DataFrame({'Investors': ['Investor1', 'Investor3', 'Investor4',
                                   'Investor2', 'Investor1', 'Investor5', 
                                   'Investor3', 'Investor2', 'Investor2', 'Investor6',
                                   'Investor5', 'Investor2', 'Investor4',
                                   'Investor7', 'Investor5'],
                          'Amount invested (in millions)': [12, 9, 17, 19, 20, 
                                                  11, 50, 12, 9, 4, 
                                                  22, 19, 17, 79, 18]})
```

```python
#plot the data for each investor's share
df_share.groupby(['Investors']).sum().plot(kind='pie', y='Amount invested (in millions)', figsize=(16,5), autopct='%1.0f%%')
plt.ylabel('')
plt.title('Investors share')
```
